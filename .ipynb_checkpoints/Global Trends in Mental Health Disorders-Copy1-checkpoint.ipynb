# Global Trends in Mental Health Disorders
The data used in this analysis was obtained here: https://data.world/vizzup/mental-health-depression-disorder-data


```python
import pandas as pd

excel_file = pd.ExcelFile('Mental health Depression disorder Data.xlsx')
# Get the list of sheet names
sheet_names = excel_file.sheet_names
sheet_names
```

```python
# Create a dictionary to store DataFrames for each sheet
dfs = {}

# Iterate over each sheet and store its DataFrame in the dictionary
for sheet_name in sheet_names:
    dfs[sheet_name] = excel_file.parse(sheet_name)
```

```python
df1 = dfs['prevalence-by-mental-and-substa']
df2 = dfs['depression-by-level-of-educatio']
df3 = dfs['prevalence-of-depression-by-age']
df4 = dfs['prevalence-of-depression-males-']
df5 = dfs['suicide-rates-vs-prevalence-of-']
df6 = dfs['number-with-depression-by-count']
```

## Prevalence Rates of Mental Health Conditions Across Different Years: A Country-Level Analysis
The dataset comprises records documenting the prevalence rates (%) of various mental health conditions (Schizophrenia, Bipolar disorder, Eating disorders, Anxiety disorders, Drug use disorders, Depression, Alcohol use disorders) across different years, identified by specific countries.

```python
df1
```

```python
# Display basic info about the dataset
print("Dataset Info:")
print(df1.info())

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(df1.head())

# Summary statistics
print("\nSummary Statistics:")
print(df1.describe())

# Check for missing values
print("\nMissing Values:")
print(df1.isnull().sum())
```

The dataset comprises 6468 entries, where each entry represents data for a specific entity, year, and disorder. The "Entity" column contains non-null object values, representing the name of the country, while the "Code" column has non-null object values representing the country's code. 

From the summary statistics, we observe variations in disorder prevalence across the years, with mean percentages ranging from approximately 0.21% for schizophrenia to about 3.99% for anxiety disorders.

The code variable is the only one with missing values

```python
from scipy import stats

# Define a threshold for outliers
threshold = 3

# Calculate z-scores for each numeric column
z_scores1 = stats.zscore(df1.select_dtypes(include=['number']))

# Find outliers
outliers1 = df1[(z_scores1 > threshold).any(axis=1)]

# Display outliers
print("\nOutliers:")
print(outliers1)
```

```python
import matplotlib.pyplot as plt

# Filter out non-numeric columns
numeric_df1 = df1.select_dtypes(include='number')

# Remove the 'Year' column
numeric_df1 = numeric_df1.drop(columns=['Year'])

# Create a box plot
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.boxplot(numeric_df1.values, labels=numeric_df1.columns)  # Plot box plots for each numeric column
plt.title('Box Plot of Mental Disorders')  # Set the title of the plot
plt.xlabel('Disorder Type')  # Set the label for the x-axis
plt.ylabel('Percentage')  # Set the label for the y-axis
plt.grid(True)  # Enable grid (optional)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()  # Show the plot
```
